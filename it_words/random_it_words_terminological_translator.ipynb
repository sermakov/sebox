```python
import time
import os
import pandas as pd
import google_trans_new
import requests
from bs4 import BeautifulSoup
from docx import Document

#google_trans_new.__version__
```

## Сбор данных с сайта

```python
alphabet = []
for i in range(65, 91):
    alphabet.append(chr(i + 32))

#alphabet
```

```python
urls = []
for l in alphabet:
    urls.append('https://www.consp.com/it-terminology-dictionary-' + l)
    
#urls
```

```python
bundle = []
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    table = soup.find('table', class_='ss-text')
    #print(table)
    #table_body = table.find('tbody')
    #print(table_body)
    
    rows = table.find_all('tr')
    #print(rows)
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        bundle.append([ele for ele in cols if ele])
        #bundle.append(row.text)# Get rid of empty values
          
display(bundle)
```

```python
to_save_bundle = pd.DataFrame(data=bundle).iloc[:,:3].dropna()
to_save_bundle.columns = ['abbreviation', 'term', 'definition']

for col in to_save_bundle.columns:
    to_save_bundle[col] = to_save_bundle[col].str.replace('\n', ' ')

#for col in to_save_bundle.columns[1:len(to_save_bundle.columns)]:
    #to_save_bundle[col] = to_save_bundle[col].str.lower()
    
to_save_bundle
```

## Перевод

```python
translator = google_trans_new.google_translator()
```

```python
translated_term_1 = []
for element in full_bundle['term'].head(125):
    #translated.append(element.upper())#translator.translate(element, lang_tgt='ru'))
    translated_term_1.append(translator.translate(element, lang_tgt='ru'))
print(translated_term_1)
```

```python
#time.sleep(60)
translated_term_2 = []
for element in full_bundle['term'].tail(125):
    #translated.append(element.upper())#translator.translate(element, lang_tgt='ru'))
    translated_term_2.append(translator.translate(element, lang_src='en', lang_tgt='ru'))
print(translated_term_2)
```

```python
translated_def_1 = []
for element in full_bundle['definition'].head(125):
    #translated.append(element.upper())#translator.translate(element, lang_tgt='ru'))
    translated_def_1.append(translator.translate(element, lang_tgt='ru'))
print(translated_def_1)
```

```python
#time.sleep(60)
translated_def_2 = []
for element in full_bundle['definition'].tail(125):
    #translated.append(element.upper())#translator.translate(element, lang_tgt='ru'))
    translated_def_2.append(translator.translate(element, lang_src='en', lang_tgt='ru'))
print(translated_def_2)
```

```python
translated_term_1.extend(translated_term_2)
translated_def_1.extend(translated_def_2)
```

```python
full_bundle['russian_term'] = translated_term_1
full_bundle['russian_term'] = full_bundle['russian_term'].str.lower()

full_bundle['russian_definition'] = translated_def_1
full_bundle['russian_definition'] = full_bundle['russian_definition'].str.lower()
```

```python
full_bundle = full_bundle.sort_values(by='term')
full_bundle = full_bundle.reset_index(drop=True)
full_bundle['number'] = full_bundle.index + 1

sorted_bundle = pd.DataFrame()
sorted_bundle[['Номер', 'Англоязычный термин', 'Англоязычное определение', 'Русскоязычный термин', 'Русскоязычное определение']] = full_bundle[['number', 'term', 'definition', 'russian_term', 'russian_definition']]

sorted_bundle
```

## Создание документа

```python
document = Document()
style = document.styles['Normal']
font = style.font
font.bold = False
```

```python
p = document.add_paragraph('Англо-русский словарь научных терминов')
```

```python
t = document.add_table(sorted_bundle.shape[0]+1, sorted_bundle.shape[1])

# add the header rows
for j in range(sorted_bundle.shape[-1]):
    t.cell(0,j).text = sorted_bundle.columns[j]

# add the rest of the data frame
for i in range(sorted_bundle.shape[0]):
    for j in range(sorted_bundle.shape[-1]):
        t.cell(i+1,j).text = str(sorted_bundle.values[i,j])
```
